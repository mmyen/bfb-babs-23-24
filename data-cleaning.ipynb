```python
import pandas as pd
from date_completeness import calculate_completeness
```

## Import and clean data

```python
all_pcs = pd.read_csv("point-count-periods.csv", nrows = 796)
all_pcs = all_pcs.drop(labels = ['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16'], axis = 1)
all_pcs
```

```python
def temperature_cleaner(temp):
    if type(temp) != str or len(temp) != 2:
        return -100
    return int(temp)
```

```python
valid_pcs = all_pcs[all_pcs['Date'] != 'error-no data']
valid_pcs['PC #'] = pd.to_numeric(valid_pcs['PC #'])
valid_pcs['Temperature'] = valid_pcs['Temperature'].apply(temperature_cleaner)
valid_pcs['Start_Datetime'] = pd.to_datetime(valid_pcs['Date'] + ' ' + valid_pcs['Start Time'])
valid_pcs['End_Datetime'] = pd.to_datetime(valid_pcs['Date'] + ' ' + valid_pcs['End Time'])
valid_pcs
```

```python
date_completenesses = valid_pcs.groupby('Date').apply(lambda x: calculate_completeness(x, x['Date'].iloc[0]))
completeness_dict = dict(date_completenesses)
```
